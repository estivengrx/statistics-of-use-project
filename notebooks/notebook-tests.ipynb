```python
df = pd.read_csv("D:/Estiven/Datos/Proyectos/statistics-of-use-project/data/tests/'app-usage-time'.csv", delimiter=";")
df = df.set_index('date', drop=True)
```

```python
def convertir_a_duracion(tiempo):
    duracion = pd.Timedelta(0)
    partes = tiempo.split()
    for parte in partes:
        if parte[-1] == 'h':
            duracion += pd.Timedelta(hours=int(parte[:-1]))
        elif parte[-1] == 'm':
            duracion += pd.Timedelta(minutes=int(parte[:-1]))
        elif parte[-1] == 's':
            duracion += pd.Timedelta(seconds=int(parte[:-1]))
    return duracion

df = df.applymap(convertir_a_duracion)
path = r'D:/Estiven/Datos/Proyectos/statistics-of-use-project/data/tests'
df.to_csv(path + 'transformed_data.csv', index=True)
```

```python
import pandas as pd

# Crear el DataFrame con los datos en formato JSON
df = pd.DataFrame({
    "Amazon Shopping": ["0s", "0s", "0s", "0s", "0s", "0s", "0s"],
    "Android System": ["4s", "4s", "0s", "9s", "1s", "12s", "0s"],
    "Biblioapp UdeA": ["0s", "0s", "0s", "0s", "0s", "0s", "0s"],
    "Bing": ["0s", "0s", "0s", "0s", "0s", "0s", "0s"],
    "Calculator": ["3m 1s", "1m 34s", "0s", "0s", "31s", "6m 40s", "8s"],
    "Calculator_1": ["0s", "1h 3m 2s", "5h 18m", "0s", "0s", "0s", "0s"],
    "Calendar": ["1m 30s", "0s", "0s", "0s", "0s", "0s", "0s"],
    "Call Management": ["0s", "1h", "0s", "0s", "0s", "0s", "0s"],
    "Camera": ["0s", "0s", "0s", "0s", "0s", "0s", "0s"],
    "Camera_2": ["15s", "8m 14s", "1m 1s", "9m 44s", "3m 48s", "7m 53s", "0s"]
})

# Recorrer cada columna del DataFrame
for col in df.columns:
    # Convertir la columna en formato de duración
    duraciones = []
    for tiempo in df[col]:
        duracion = pd.Timedelta(0)
        partes = tiempo.split()
        for parte in partes:
            if parte[-1] == 'h':
                duracion += pd.Timedelta(hours=int(parte[:-1]))
            elif parte[-1] == 'm':
                duracion += pd.Timedelta(minutes=int(parte[:-1]))
            elif parte[-1] == 's':
                duracion += pd.Timedelta(seconds=int(parte[:-1]))
        duraciones.append(duracion)
    df[col] = duraciones

# Mostrar el DataFrame resultante
df
```
