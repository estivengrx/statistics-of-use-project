```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("'App - Usage Time$'.csv", delimiter=";")
df = df.set_index('date', drop=True)
df
```

```python
import re

def convertir_a_formato_hora(duracion):
    horas = minutos = segundos = 0

    # Utilizamos expresiones regulares para extraer las horas, minutos y segundos
    horas_match = re.search(r'(\d+)h', duracion)
    minutos_match = re.search(r'(\d+)m', duracion)
    segundos_match = re.search(r'(\d+)s', duracion)

    # Convertimos los valores extraídos a enteros si existen
    if horas_match:
        horas = int(horas_match.group(1))
    if minutos_match:
        minutos = int(minutos_match.group(1))
    if segundos_match:
        segundos = int(segundos_match.group(1))

    # Formateamos la duración en formato HH:MM:SS
    duracion_formateada = "{:02d}:{:02d}:{:02d}".format(horas, minutos, segundos)

    return duracion_formateada

# Ejemplo de uso
duracion1 = "0s"
duracion2 = "1h 30m 15s"

duracion_formateada1 = convertir_a_formato_hora(duracion1)
duracion_formateada2 = convertir_a_formato_hora(duracion2)

print(duracion_formateada1)  # Output: 00:00:00
print(duracion_formateada2)  # Output: 01:30:15
```

```python
df = df.applymap(convertir_a_formato_hora)
```
