```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/Estiven/Datos/Proyectos/statistics-of-use-project/data/processed/app_usage_time_motorola.csv", 
                 delimiter=";",
                 index_col='date',
                 parse_dates=['date'])
df.head(2)
```

```python
# Iterate over each column in the DataFrame
for col in df.columns:
    # Convert the column values to duration format
    df[col] = round(pd.to_timedelta(df[col]).dt.total_seconds() / 60, 2)

# Convert the DataFrame to duration data type
df
```
