```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/Estiven/Datos/Proyectos/statistics-of-use-project/data/processed/web_usage_time_motorola.csv", 
                 delimiter=";",
                 parse_dates=['date'],
                 encoding='utf-8')
df
```

```python
# Iterate over each column in the DataFrame
for col in df.columns:
    if col != 'date':
        df[col] = round(pd.to_timedelta(df[col]).dt.total_seconds() / 60, 2)
    else: pass

# Convert the DataFrame to duration data type
df
```

```python
df.to_csv('D:/Estiven/Datos/Proyectos/statistics-of-use-project/data/processed/web_usage_time_motorola.csv', index=False)
```
